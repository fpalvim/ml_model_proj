Limpieza de los datos

Cambio de nombre de las columnas para castellano

```python
new_column_names = {
    'Age': 'Edad',
    'Sex': 'Sexo',
    'ChestPainType': 'TipoDolorTorax',
    'RestingBP': 'PresionArterialReposo',
    'Cholesterol': 'Colesterol',
    'FastingBS': 'GlucosaEnAyunas',
    'RestingECG': 'ECGReposo',
    'MaxHR': 'FreqCardiacaMaxima',
    'ExerciseAngina': 'AnginaDeEsfuerzo',
    'Oldpeak': 'DescensoST',
    'ST_Slope': 'PendienteST',
    'HeartDisease': 'EnfermedadCardiaca'
}
```

```python
data.rename(columns=new_column_names, inplace=True)
```

se nota la presencia de valores min = 0 en las columnas Cholesterol y FastingBS, esto no puede pasar, por esto vamos a ver cuantos son estes datos y lo que podemos hacer con ellos

```python
data[data['PresionArterialReposo'] == 0]
```

en el caso de la columna RestingBP vamos a quitar dicho registro, pues no causa perdida en la analise de todo el conjunto de datos, devido a cuantidad sin relevancia

```python
data = data.loc[data['PresionArterialReposo'] != 0]
```

Vamos hacer la misma comprobacion en la columna Cholesterol

```python
data['Colesterol'].value_counts()
```

Como son muchos valores, para esta columna no se puede borrar tales datos, vamos substituir los valores 0 de esta columna por el valor de la media de los valores de la columna. Para calcular la media de esta columna, vamos calcular sin utilizar los valores que actualmente estan iguales a 0 para no sesgar los datos calculados.

```python
data['Colesterol'].replace(0, np.nan, inplace=True);
```

```python
data['Colesterol'].fillna(data['Colesterol'].mean(), inplace=True);
```
